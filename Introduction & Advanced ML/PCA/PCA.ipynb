```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Task 1

```python
df = pd.read_csv('13_36.csv', sep=',', header=None)
pca = PCA()
new_df = pd.DataFrame(pca.fit_transform(df))
expl_var = pca.explained_variance_ratio_
```

```python
### ANSWER
print("First PC coord of first object: ", new_df.loc[0,0])
print("Second PC coord of first object: ", new_df.loc[0,1])
print()
for i in range(len(expl_var)):
  print("Explained varience ration using " +str(i+1)+ " PC: ", np.sum(expl_var[:(i+1)]))
print()
print("Count the groups on plot")
plt.scatter(new_df[0], new_df[1])
plt.show()
```
