```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Task 1

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasets/8_36.csv', sep=',', index_col=False, skipinitialspace=True, header=None)
pca_c = PCA()
new_data = pd.DataFrame(pca_c.fit_transform(data))
expl_var = pca_c.explained_variance_ratio_
```

```python
### ANSWER
print("First PC value of first object: ", new_data.loc[0,0])
print("Second PC value of first object: ", new_data.loc[0,1])
print()
for i in range(len(expl_var)):
  print("Explained varience ration using " +str(i+1)+ " PC: ", np.sum(expl_var[:(i+1)]))
print()
print("Count the groups on plot")
plt.scatter(new_data[0], new_data[1])
plt.show()
```
